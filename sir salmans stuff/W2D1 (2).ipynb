```python
from time import time
def measureTime(func):
    def wrapper():
        start_time = time()
        func()
        end_time = time()
        print("excution time ", end_time - start_time)
    return wrapper
```

```python
@measureTime
def doSomeLooping():
    for i in range(30):
        print(i, end='\r')
doSomeLooping()
```

```python
@measureTime
def filerPassword():
    passwords = ["a", "abc", "ab", "alsdkfj", "a;sldkjf;laksdjfklj", "alsdkjfkjwer"]
    filtered_passwords = []
    for passsword in passwords:
        if len(passsword) >= 8:
            filtered_passwords.append(passsword)
    return filtered_passwords
```

```python
name_arr = []
age_arr = []
height_arr = []
for i in range(5):
    print(f"user # {i}")
    name = input("enter name")
    age = input("enter age")
    height = input("enter height")
    name_arr.append(name)
    age_arr.append(age)
    height_arr.append(height)
data = {'name': name_arr, 'age':age_arr, 'height':height_arr}
```

```python
with open('people.csv', 'w') as file:
    file.write(",".join(list(data.keys())) + "\n")
    for i in range(5):
        file.write(data['name'][i] + "," + data['age'][i] + "," + data['height'][i] + "\n")
```
